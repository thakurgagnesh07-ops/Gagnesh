```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df =pd.read_csv("s.csv")
```

```python
df.info
```

```python
dept_count = df['Department'].value_counts()

plt.figure(figsize=(8,5))
dept_count.plot(kind='bar')
plt.title("Number of Employees in Each Department")
plt.xlabel("Department")
plt.ylabel("Employee Count")
plt.show()
```

```python
df['Age'].value_counts().plot(kind='pie')
plt.show()
```

```python
df['Salary'].value_counts().plot(kind='pie')
plt.show()
```

```python
gender_count = df['Gender'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(gender_count,
        labels=gender_count.index,
        autopct='%1.1f%%')

plt.title("Gender Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Age'], bins=10)
plt.title("Age Distribution of Employees")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.plot(df["Department"], df["Position"])
plt.title("graph")
plt.xlabel("work")
```

```python

dept_count = df['Department'].value_counts()

plt.figure(figsize=(8,5))
dept_count.plot(kind='bar')
plt.title("Department-wise Employee Count")
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.show()

print("Department with Highest Employees:")
print(dept_count.idxmax())
```

```python

dept_count = df['Position'].value_counts()

plt.figure(figsize=(8,5))
dept_count.plot(kind='bar')
plt.title("Position-wise Employee Count")
plt.xlabel("Position")
plt.ylabel("Number of Employees")
plt.show()

```

```python

avg_salary = df.groupby('Department')['Salary'].mean()

plt.figure(figsize=(8,5))
avg_salary.plot(kind='bar')
plt.title("Average Salary by Department")
plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.show()
```
